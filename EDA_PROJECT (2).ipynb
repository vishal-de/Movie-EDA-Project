```python
num_col=['BudgetUSD','US_BoxOfficeUSD','Global_BoxOfficeUSD','Opening_Day_SalesUSD', 'One_Week_SalesUSD', 'IMDbRating','RottenTomatoesScore', 'NumVotesIMDb', 'NumVotesRT',]
```

```python
import matplotlib.pyplot as plt

for col in num_col:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=50, density=True, alpha=0.6)
    
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    
    plt.show()
```

```python
df.columns
```

```python
cat_col=['Genre','Country','Director',
       'LeadActor']
```

```python
import matplotlib.pyplot as plt

for col in cat_col:
    plt.figure(figsize=(8,4))
    
    df[col].value_counts().head(20).plot(kind='bar')
    
    plt.title(f"Top 20 {col} Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

plt.scatter(
    df['BudgetUSD'],
    df['Global_BoxOfficeUSD'],
    alpha=0.4,
    s=30,
    color='blue',
    edgecolor='black',   # this adds the border
    linewidths=0.5
)

plt.title('Budget vs Global Box Office')
plt.xlabel('BudgetUSD')
plt.ylabel('Global_BoxOfficeUSD')

plt.xscale("log")
plt.yscale("log")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

plt.scatter(
    df['IMDbRating'], 
    df['RottenTomatoesScore'], 
    alpha=0.3
)

plt.title('IMDb Rating vs Rotten Tomatoes Score')
plt.xlabel('IMDbRating')
plt.ylabel('RottenTomatoesScore')

plt.tight_layout()
plt.show()
```

```python
genre_year_stats=df.groupby(['Genre', 'ReleaseYear']).agg({'Global_BoxOfficeUSD':'mean','BudgetUSD':'mean','IMDbRating':'mean','RottenTomatoesScore':'mean'}).reset_index()
genre_year_stats
```

```python
top_actor = (
    df.groupby("LeadActor")["IMDbRating"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

print(top_actor)
```

```python
top_director = (
    df.groupby('Director')["IMDbRating"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

print(top_director )
```

```python
max_sales_row = df.loc[df['Global_BoxOfficeUSD'].idxmax()]
movie_name = max_sales_row['Title']
sales = max_sales_row['Global_BoxOfficeUSD']
print("Highest Grossing Movie:", movie_name)
print("Global Box Office Sales:", sales)
```

```python
min_sales_row = df.loc[df['Global_BoxOfficeUSD'].idxmin()]
movie_name = min_sales_row['Title']
sales = min_sales_row['Global_BoxOfficeUSD']
print("lowest Grossing Movie:", movie_name)
print("Global Box Office Sales:", sales)
```

```python
import matplotlib.pyplot as plt


action_data = genre_year_stats.query("Genre == 'Action'")

plt.figure(figsize=(15, 10))

plt.plot(
    action_data['ReleaseYear'],
    action_data['Global_BoxOfficeUSD'],
    marker='o'
)

plt.title('Average Global Box Office by Genre Over Years')
plt.xlabel('Release Year')
plt.ylabel('Global Box Office (USD)')

plt.grid(True)
plt.show()
```
